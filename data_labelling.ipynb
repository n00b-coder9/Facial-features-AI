```python
import os
import pandas as pd
## put here the full path of the generated file of the annotation tool, e.g.: /Users/jim/Desktop/AU_video.txt
file1_name = 'C:/Users/Faiz/Downloads/AU_video.txt'
file2_name = 'D:/tutors_point/courework annotation tool/Attribute_annotation_tool/examples/image_names.txt'
with open(file1_name) as fil:
 content = fil.readlines()
with open(file2_name) as fil:
 content2 = fil.readlines()
outputs = []
headers = ["image_name","wrinkles","freakles","glasses","hair_color","hair_top","no face shown or not human"]
for elem,elem1 in zip(content,content2) :
 if 'N/A' in elem:
   continue
 else:
   line = elem.split(',')
   line1 = elem1.split('/')
   outputs.append([line1[5].split('\n')[0],int(line[1]),int(line[3]),int(line[5]),int(line[7]),int(line[9]),int(line[11])])
pd.DataFrame(outputs).to_csv("annoitations.csv", header = headers , index = None )
print(outputs)
# always the outputs correspond to (in that order): wrinkles, freakles, glasses, hair color, hair top, no face shown or not human
     
```
